## Gender Analysis

```python
# Taking the number of total males and females and converting it into a dataframe
data=pd.DataFrame({'Ratio': [len(df[df["Gender"]=='M']),len(df[df['Gender']=='F'])]},index=['Male','Female'])
```

```python
data
```

```python
# Displaying the gender ratio in percentage
data.plot.pie(y='Ratio',autopct='%.1f',title='Gender Ratio')
```

### Other simple way to do it:

```python
df.groupby('Gender').size().plot(kind='pie',autopct='%.1f',title='Gender Ratio')
```

```python
# Displaying average spendage of each of the gender
df.groupby('Gender').mean()['Purchase'].plot(kind='pie',autopct="%.1f",title='Average Spendage')
```

```python
# Displaying total amount spendage genderwise
df.groupby('Gender').sum()['Purchase'].plot(kind='pie',autopct='%.1f',title='Total Spendage genderwise')
```

## Combining gender and marital status for analysis

### There are 4 categories:
### M0 (Male and unmarried)
### M1 (Male and married)
### F0 (Female and unmarried)
### F1 (Female and married)

```python
l=[] # Creating a list
for i in range(len(df)):
    l.append(df['Gender'][i]+" "+str(df['Marital_Status'][i])) # This will append gender and marital status in the list
df['Marital_Gender']=l # Making a new column in the df
```

```python
df.head()
```

```python
sns.countplot(df['Marital_Gender'])
```

## Unmarried Males tend to buy the most during the black friday sale

```python
sns.countplot(x=df['Age'],hue=df['Marital_Gender'])
```

## Unmarried males of the  age group 26-35 tend the buy the most

```python
sns.countplot(x=df['Product_Category_1'],hue=df['Marital_Gender'])
```

```python
sns.countplot(x=df['City_Category'],hue=df['Marital_Gender'])
```
